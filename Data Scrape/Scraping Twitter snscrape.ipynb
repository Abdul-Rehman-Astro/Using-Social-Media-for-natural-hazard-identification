```python
type(scraper)
```

```python
tweet_data_with_user_location=[]
for i,tweet in enumerate(scraper.get_items()): 
    tweet_list = [
        tweet.date,
        tweet.id,
        tweet.content,
        tweet.user.username,
        tweet.likeCount,
        tweet.retweetCount,
        tweet.user.location  # Add the user location to the tweet_list
    ]
    tweet_data_with_user_location.append(tweet_list)
    if i==10000:
        break
    
```

```python
# tweet_df = pd.DataFrame(tweet_data, columns=['date', 'id', 'content', 'username', 'like', 'retweet'])
tweet_flood_df = pd.DataFrame(tweet_data_with_user_location, columns=['date', 'id','content','username', 'like','retweet', 'user_location'])
```

```python

tweet_flood_df['date'] = pd.to_datetime(tweet_flood_df['date']).dt.date
```

```python
tweet_flood_df.to_excel("twitter_with_user_location.xlsx",index= False)
```
